```python
df = pd.read_csv('/Users/jihyeon/Documents/K_Semi/bmiset.csv')
```

```python
df.head()
```

```python
import math
#BMI 정리 함수
def bmirange(x):
    x=math.trunc(x)
    if x>= 45:
        x=45
    return x
```

```python
df['BMI_range']=df['BMI'].apply(bmirange)
```

```python
df.head()
```

```python
# Yes / No 정리
def yesno(thing):
    if thing.lower()=='yes':
        thing=1
    else:
        thing=0
    return thing
```

```python
df['HeartDisease']=df['HeartDisease'].apply(yesno)
df['Stroke']=df['Stroke'].apply(yesno)
df['Diabetic']=df['Diabetic'].apply(yesno)
df['KidneyDisease']=df['KidneyDisease'].apply(yesno)
df['PhysicalActivity']=df['PhysicalActivity'].apply(yesno)
```

```python
df.head()
```

```python
#bmi-age 피벗테이블:당뇨
bmi_pv=pd.pivot_table(df,values='Diabetic',index='AgeCategory',columns='BMI_range')
```

```python
bmidp=bmi_pv*100
```

```python
sns.set(rc = {'figure.figsize':(24,8)})
```

```python
ax=sns.heatmap(bmidp,annot=True,fmt='.1f',cmap='coolwarm', linewidths=.5)
```

```python
ax.get_figure().savefig("/Users/jihyeon/heatmap_Diabetic.png")
```

```python
#bmi-age 피벗테이블:심장질환
bmi_pvh=pd.pivot_table(df,values='HeartDisease',index='AgeCategory',columns='BMI_range')
```

```python
bmiht=bmi_pvh*100
```
